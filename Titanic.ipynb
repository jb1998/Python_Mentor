```python
def getNumber(s):
    if (s == 'male'):
        return 1
    else:
        return 0
    
titanic['gender'] = titanic['Sex'].apply(getNumber)
test['gender'] = titanic['Sex'].apply(getNumber)
```

```python
del titanic['Sex']
del test['Sex']
titanic.head()
```

```python
titanic.Embarked.value_counts()
```

```python
titanic.Embarked.fillna('S',inplace=True)
titanic.Embarked.value_counts()
```

```python
titanic.describe()
```

```python
age_average=titanic.Age.sum()/(891-177)
titanic.Age.fillna(age_average,inplace=True)
age_average2=test.Age.sum()/(418-86)
test.Age.fillna(age_average2,inplace=True)
fare_average=test.Fare.sum()/417
test.Fare.fillna(fare_average,inplace=True)

test.isnull().sum()
```

```python
# test = (test - test.mean()) / test.std()
target=titanic.Survived.copy()
```

```python
del titanic['Survived']
# target
```

```python
from sklearn import tree
```

```python
def getNumber2(s):
    if (s == 'S'):
        return 1
    else:
        return 0
def getNumber3(s):
    if (s == 'C'):
        return 1
    else:
        return 0
def getNumber4(s):
    if (s == 'Q'):
        return 1
    else:
        return 0
    
titanic['is_S'] = titanic['Embarked'].apply(getNumber2)
titanic['is_C'] = titanic['Embarked'].apply(getNumber3)
titanic['is_Q'] = titanic['Embarked'].apply(getNumber4)
del titanic['Embarked']

test['is_S'] = test['Embarked'].apply(getNumber2)
test['is_C'] = test['Embarked'].apply(getNumber3)
test['is_Q'] = test['Embarked'].apply(getNumber4)
del test['Embarked']
```

```python
test.head()
```

```python
clf = tree.DecisionTreeClassifier()
# clf = clf.fit(iris.data, iris.target)
```

```python
clf.fit(titanic, target)
```

```python
predictions=clf.predict(test)
```

```python
prediction = pd.DataFrame(predictions, columns=['predictions']).to_csv('titanic_prediction.csv')
```
